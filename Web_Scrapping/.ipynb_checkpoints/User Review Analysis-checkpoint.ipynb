```python
reviews = pd.read_csv('~/vulpix/data/all_reviews.csv',sep='\t',encoding='utf-8', index_col=0)
reviews = reviews.drop('reviews', 1)
```

```python
name_avg = reviews.groupby('name').overall.mean()
```

```python
# sort animes with more than one review
anime_vc = reviews.anime.value_counts()
sub_reviews = reviews[reviews.anime.isin(anime_vc[anime_vc > 10].index)]
```

```python
def compute_supports(df):
    uuids=df.name.unique()
    rests = df.groupby('anime')['name'].unique()
    bdict={}
    for e,v in zip(rests.index.values, rests.values):
        bdict[e] = np.array([item in v for item in uuids])
    anime=bdict.keys()
    supports=[]
    for i,a1 in enumerate(anime):
        for j,a2 in enumerate(anime):
            if  i < j:
                supmask = (bdict[a1] & bdict[a2])
                common_reviewers = np.sum(supmask)
                supports.append(common_reviewers)
    print "mean support",np.mean(supports), "median support", np.median(supports), "max support", np.max(supports)
    return supports, bdict
```

```python
s,d = compute_supports(sub_reviews)
```

```python
plt.hist(s, bins=np.arange(1,30,1), alpha=0.5, label="support");
c=sub_reviews.groupby('name')['anime'].count()
print "User review counts mean and median",np.mean(c), np.median(c)
plt.hist(c, bins=np.arange(1,30,1), alpha=0.5, label="user reviews");
plt.legend();
```
